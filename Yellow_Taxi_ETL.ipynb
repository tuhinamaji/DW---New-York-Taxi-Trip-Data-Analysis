```python
def download_file(url, save_path):
    response = requests.get(url, stream=True)
    if response.status_code == 200:
        with open(save_path, 'wb') as f:
            f.write(response.content)
        print(f"Downloaded: {url}")
    else:
        print(f"Failed to download: {url}")

def download_taxi_data(year_list, months, base_url, save_dir):
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)

    for year in year_list:
        for month in months:
            month_padded = str(month).zfill(2)  # Ensure month is two digits
            file_name = f"yellow_tripdata_{year}-{month_padded}.parquet"
            url = f"{base_url}{file_name}"
            save_path = os.path.join(save_dir, file_name)
            download_file(url, save_path)


months = range(1, 13)
base_url = 'https://d37ci6vzurychx.cloudfront.net/trip-data/'

download_taxi_data(year_list, months, base_url, save_dir)
```

### DEALING WITH NANS

## DEALING WITH SOR ANOMOLIES

```python
df_main['trip_time']=df_main.tpep_dropoff_datetime - df_main.tpep_pickup_datetime
```

```python
df_main['trip_time_minutes'] = df_main['trip_time'].dt.total_seconds() / 60
```

Only rows with positive trip times

```python
df_main = df_main[df_main.trip_time_minutes > 0]
```

```python
database_name = 'bi_and_dw'
```

```python
from sqlalchemy import create_engine

engine = create_engine(f'mysql+mysqlconnector://{username}:{password}@{hostname}')

with engine.connect() as conn:
    # Check if database exists
    db_exists = conn.execute(f"SHOW DATABASES LIKE '{database_name}';").fetchone()

    # Create database if it does not exist
    if not db_exists:
        conn.execute(f"CREATE DATABASE {database_name};")
        print(f"Database {database_name} created.")
    else:
        print(f"Database {database_name} already exists.")

# After ensuring the database exists, reconnect to the specific database
engine = create_engine(f'mysql+mysqlconnector://{username}:{password}@{hostname}/{database_name}')
```
